# 1\. Sarrera

Segurtasun kontuekin hasi aurretik sare mailan egiten diren oinarrizko konfigurazioak ikusiko ditugu, aurrerago horiek ulertzea ezinbestekoa izango baita segurtasuna bermatzeko hartu behar ditugun neurriak artzeko erabaki bat edo besteren oinarri izango baitira.

Unitate didakatiko honetan nagusiki bi kontu landuko ditugu: Bideratzaileak eta VLANak. Horretarako, batetik, Cisco teknologian horien inguruko atazak ezberdinak nola burutzen diren ikusiko dugu Packet Tracer sare simulatzailearen bidez eta, bestetik, L3 motako Mikrotik Switch/Router baten bidez kontzeptu berdinak nola gauzatzen diren ikusiko dugu.

Hemen landuko digutun kontzeptuak, besteak beste:

- Sareko gailuen atzipen aukera ezberdinak
- Gailuen konfigurazioen kudeaketa
- Bideratze estatikoa
- Bideratze dinamikoai
- NAT mota ezberdinak
  - PAT
  - Port Fordwarding
- Trafikoa filtratzea
- Erabiltzaileen kudeaketa sareko gailuetan
- VLANen kontzeptuak eta konfigurazio aukerak
- Izidenteen diagnosirakok aukerak


# 2\. Routerretara atzipena

## 2.1. Atzipenaren oinarriak

Sareko beste gailu kudegarri gehientsuenetara bezala, hainbat konfigurazio egiteko eskainzen dituzte routerrek atzipen bide ezbedinak.

Bost mota nagusi ezberdin daude routerrak, eta oro har sareko gailuak, atzitzeko:

### **2.1.1 Serie kable bidezko atzipena**

Router profesionalek eskaintzen duten oinarrizko konexioa da eta besterik aktibatuta ez dagoenean ere erabili daiteke. Serie konexioa eskaintzen duen ataka berezi bat eskaintzen du gailuak eta bertara espreski horretarako bide ematen duten kableen bitartez konektatu eta serie konexioak burutu ahal izateko programa baten bidez lan egin dezakegu.

### **2.1.2 SSH bidezko atzipena**

Gailuan aukera hori gaituta baldin badaukagu eta beharrezko baimenak dituen erabiltzailea baldin badugu. Inguru grafikoan ematen diren aukerak baino gehiago ematen dira kasu askotan eta horrez gain, konfiguraziorako komandoak menperatzen diren kasuan, konfigurazioak azkarrago burutzen dira eta automatizaziorako scriptak konfiguratu daitezke kasu askotan.

### **2.1.3 Web bidezko atzipena**

Gailuan aukera hori gaituta web bidezko interfaze baten bide atzitu daitezke routerrak eta hor eskaintzen diren aukera (kasu batzuetan ez guztiak) baliatu daitezke konfigurazioak gauzatzeko. Konfiguazioa txikiak egiteko edo ezagutza handia ez denean erabiltzeko gomendagarria da.


### **2.1.4 Bezero software bidezko atzipena**

Marka ezberdinek beraien produktuak kudeatzeko softwarea eskaitzen duten. Gailutik aparte exekutzen den inguru grafiko honen bidez funtzio gehienak konfiguratu daitezke. Kasu askotan aldaketa guztiak software horretan bertan egiten dire eta gero kargatzen dira dagokion gailuan. Modu honetan (beste aukeretan ere posible bada ere) konfigurazioen bertsioen kontrola egitea ahalbidetzen da.


### **2.1.5 Kontrolatzaile zentralizatu bidez**

Aurreko aukeraren eboluzioa bat da. Kasu honetan aparteko software bidez gailuak kontrolatu ordez, gailuak lotura ezarria duten kontrolatzaile zentralizatu bakarraren bidez kudeatu daitezke. Kasu hauetan ere marka bakoitzak bere baliabideak eskaintzen ditu. Maila lokalean instalatu daitezkeen kontrolatzaileak edo zeregin batzuetarako hodeian izan daitezkeenak aurkitu daitezke.

## 3.4. Helbideen itzulpena: NAT eta PAT

Atal honetan SNATa (Source NAT) edo jatorriaren gaineko NATa (Network Address Translation) soilik hartuko dugu kontutan.

- NAT estatikoa: helbide pribatu bakoitzeko helbide publiko bat egokitzen zaio. Hau ez da oso eraginkorra SNAT egiteko, izan ere, horrek gure saretik kanpora komunikazioa behar duen gailu bakoitzeko ip publiko bat behar baitugu.
- NAT dinamikoa: helbide publiko kopurua txikiagoa da eta barneko ip bakoitzari egokituko zaio beharraren arabera. Modu honetan aldi berean kanpoko baliabideak erabiliko dituzten gailuak haina ip publiko behar ditugu
- NAT gainkargatua edo PAT: Kasu honetan IP publiko bakarra ere nahikoa da, egiten den mapeoa ez baita soilik IP mailan egiten baizik eta IP eta portu mailan. Hau da NAT mota ohikoena:

## 3.6. Trafikoa filtratzea ACLak erabiliz

Ciscoren Routerretan, beste marka askorenetan bezala, trafikoa filtratu egin daiteke eta hori ACLen (Access Control List) bidez egiten da. Horretarako filtroa ezarri nahi dugun interfazeari ACL bat egikotu behar dugu eta ACL zerrenda horretan bete behar diren arauak gehitu.

ACL zerrendak oro har (beste aukera batzuk badaude ere) zenbaki bidez identifikatzen dira eta Ciscon bi motakoak daude: arruntak (edo oinarrizkoak) eta hedatuak. Arruntak paketearen jatorri helbidean oinarritzen dira eta hedatuek jatorria, helburua, protokolo mota eta protokolo identifikatzailea hartu ditzakete kontutan besteak beste.

ACL baten identifikatzaile gisa aukeratzen dugun zenbakiaren arabera hedatua dela edo ez adierazten dugu: 100 eta 199 artekok eta 2000 eta 2699 artekoak hedatuak dira.

Beste berezitasun bat, interfazeari egokitzerakoan interfaze horretatik sartzen den trafikoarentzat ACL bat egokitu dezakegula eta irteten denarentzat beste bat.

Aurreko ariketako emaitzatik abiatuta (Arik 1.3) eraikiko dugun adibide berri batekin (Adib 1.4) ikusiko dugu nola konfiguratu daitezken:

![image](img/cisco-adib1.4.png) 

- Hasteko, ikastetxeko routerraren kanpoaldetik barrurantza sartzen dena kontrolatzea interesatzen zaigu:

![image](img/cisco-ikastetxea-ACL.png) 

    interface gigabitEthernet 0/0/1
    ip access-group 100 in
    exit

- Lehen arau edo filtroa gehituko dugu. Kanpotik datorren eta jatorrian helbide pribatua (soilik 192.168.0.0/16) duen trafikoa ez dugu onartuko:

![image](img/cisco-ikastetxea-ACL-denyall.png) 


    access-list 100 deny tcp 192.168.0.0 0.0.255.255 any
    access-list 100 deny icmp 192.168.0.0 0.0.255.255 any

- Onartu ezarritako tcp konexioen parte diren paketeak eta ping erantzunak:

![image](img/cisco-ikastetxea-ACL-buelta.png) 

    access-list 100 permit tcp any 8.0.0.1 0.0.0.0 established
    access-list 100 permit icmp any 8.0.0.1 0.0.0.0 echo-reply

- Onartu DMZko zerbitzarirako http eskaria:

![image](img/cisco-ikastetxea-ACL-http.png) 

    access-list 100 permit tcp any 8.0.0.1 0.0.0.0 eq www

# 5\. VLANak

Askotan sareak segmentatzea (azpisare bereizi ezberdinetan zatitzea) interesatzen zaigu, errendimendua, txukuntasuna eta segurtasuna ezartzeko. Lehen ikuspegi baten VLANek switchak birtualki guztiz banatu egiten dituzte beraz VLANetan bereizten dugun switch bat bi switch erabiltzearen antzekoa da. Modu horretan interpretzen badugu, lotura tronkaletan kable batetetik kable virtual bat baino gehiago doala ulertuko genuke, beraz kable bat baino gehiagorekin ordezkatuko litzateke guztia fisikoki bereizita egingo bagenu.

## 5.1. VLANak Cisco switchetan

Packet tracerreko adibide (Adib 1.7) batekin ikusiko dugu zertarako behar diren VLANak eta nola inplementatu daitezkeen Ciscon. Demagun ondoko sare fisikoa daukagula:

![image](img/cisco-vlan-fisikoa.png)

Guk ondoko era logikoan aritzea nahiko genuke:

![image](img/cisco-vlan-logikoa.png)

Logika hori ezartzeko, VLANak erabiliko ditugu eta beraz hiru sare horietako bakoitza VLAN batean egongo da. Adibidez Ikasleak VLAN 10, Irakasleak VLAN 20 eta sistemak VLAN 30. Aurrerago aipatuko badugu ere kontuan hartu behar da VLANaren zenbakia ia guztiz askeki erabili dezakegu eta horren arabera bereiziko dira sare logikoak.

Ikusten ari garen adibidean arreta jartzen badugu, ikus dezakegu tarte batzuetan VLANek ez soilik gailu nagusi (switcha) partekatzen dutela, baizik eta kable bat eta hau konektatuta dagoen portuak ere partekatzen dituztela. Honekin lotuta, bi VLAN portu bereizi ditzakegu:

- VLANera sarrera portuak: Ciscon Access deitzen dira eta beste ingurune batzuentan Untagged bezala identifikatzen dira. Bereziketa hasi behar den switchera oraindik bereizketarik ez duen gailu edo sarea konektatzen den portu mota da. Beste era batera esanda, hortik sartuko diren paketeek, ordura arte ez dute VLAN etiketarik (Unttaged daude) eta orduan identifikatzen (taggeatzen( dira VLAN horretan). Informazioa hortik sartu ordez atera egin behar denean kontrakoa egingo da.
- VLANen hedapen portuak: Ciscon Trunk deitzen dira eta beste ingurune batzuetan Tagged bezala identifikatzen dira. Bereizketa mantendu egiten da portu hauetan eta beraz hortik doan informazioa dagokion VLAN zenbakiarekin etiketatuta (taggeatuta) mantentzen da, bai barrurantz bai kanporantz. Normalean hau VLAN ezberdinak daramatzan konexio puntua izaten da.

Honen arabera, gure adibidean ondoko eran konfiguratu beharko lirateke Access (A) edo Trunk (T) portuak:

## 5.2. VLANak Cisco routerretan

Lotune hori routerrentan (edo bideratzea gauzatzeko gai den gailuetan: L3 switchak, Suheziak …) aurkitzen da eta hurrengo adibidearen (Adib 1.8) bidez ikusiko dugu nola konfiguratzen den:

Rooterreko lotura ezartzeaz gain, adibidea aprobetxatu eta muga batzuk ere jarriko ditugu ACL bidez. Ondoko parametroak hartuko ditugu kontuan:

  - Ikasleen azpisarea: 192.168.10.0/24. Internetera atzipena izango du, baina beste azpisareetara ez.
  - Irakasleen azpisarea: 192.168.20.0/24. Internetera eta beste bi azpisareetara atzipena izango du.
  - Sistemak azpisarea: 192.168.30.0/24 Beste bi aspisareetara atzipena izango du, baina internetera ez.

Hortik abiatuta lehendabizi switcharen eta routerraren arteko lotura konfiguratuko dugu:

- Switchean:

        interface GigabitEthernet0/2
        switchport mode trunk
        exit

  - VLANak izendatu behar ditugu

        vlan database
        vlan 10 name Ikasleak
        vlan 20 name Irakasleak
        vlan 30 name Sistemak
        exit

        interface GigabitEthernet0/0/0

        exit
        interface GigabitEthernet0/0/0.10

        ip address 192.168.10.254 255.255.255.0
        exit
        interface GigabitEthernet0/0/0.20

        ip address 192.168.20.254 255.255.255.0

## 5.4. VLAN Protokoloak

VLANekin lan egiten dugunean sarean bidaiatzen duten tramak aldatu egiten dira. IEEE erakundeak IEEE 802.1q (dot1q bezala ere ezagutzen dena) estandarra definitu zuen VLANei zegozkien sareko parametroak finkatzeko eta tartean jatorrizko trama aldatzen duen dot1q enkapsulazioa ere definitu zuen. Enkapsulazio deitzen den arren, berez ez da horrelakorik egiten traman informazioa gehigarria txertatu baizik.

Badaude estandar hori bete ordez enkapsulazioa beste era batera egiten duten teknologiak ere, baina gehienek, bateragarritasunagatik, dot1q aukera ere beti ematen dute: adibidez Ciscok badu dot1q enkapsulazioaren alternatiba den ISL protokoloa, baina normalena dot1q erabiltzea da ciscoko gailuekin ere.

Hala ere badaude enkapsulazioatik aparteko zereginetarako markek beraien kontura erabiltzen dituzten protokoloak: adibidez, Ciscok, gailu ezberdinetara VLANen konfigurazioari buruzko informazioa hedatzeko (konfigurazioa errazte aldera) dot1q barruan definituta dagoen MVRP erabili ordez, beraiek diseinatutako VTP protokoloa erabiltzen da gehienetan.
