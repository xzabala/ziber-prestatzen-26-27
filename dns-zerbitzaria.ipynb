# Webmin eta BIND instalazioa Ubuntu 24.04ean

Fitxategi hau Ubuntu 26.04 sistema batean Webmin eta BIND (DNS zerbitzaria) nola instalatu eta konfiguratu azaltzen du.
Hau gelako DNS zerbitzaria montatzeko erabili dugu.

## 0. Ubuntu instalatu, SSH zerbitzaria instalatu eta sare konfigurazioa egiaztatu

1. Ubuntu 26.04 instalatu eta sistema martxan jarri.
2. SSH zerbitzaria instalatu eta abiarazi:

````python
sudo apt update
sudo apt install -y openssh-server
sudo systemctl enable --now ssh
````

3. SSH zerbitzuaren egoera egiaztatu:

````python
sudo systemctl status ssh
````

4. Sare konfigurazioa egiaztatu nahi baduzu, zure IP helbidea eta sareko interfazeak kontsultatu (netplan aldatuta dagoe onezkero):

````python

ip a
ip route
````

5. SSH konexioa probatu beste makina batetik edo bertatik bertara erabiliz:

- Bestetik, `ssh <erabiltzailea>@<zure-ip-helbidea>`.
- Bertatik bertara probatzeko: `ssh localhost` edo `ssh <erabiltzaile-izena>@127.0.0.1`.

6. Ubuntun `ufw` martxan dagoen kasuan, ireki 22. portua SSH-rako:

````python
sudo ufw allow 22/tcp
sudo ufw enable
````

7. Proba arrakastatsua izan bada, SSH zerbitzariak ondo dabilela baieztatu dezakezu.

## 1. Sisteman eguneratzea

Terminalean jarraitu kode hauekin:

````python
sudo apt update
sudo apt upgrade -y
````

Honek erabilgarri dauden pakete berrienak jarriko ditu eta sistema eguneratua izango da.

## 2. Webmin instalatzea

Webmin web-interfaze grafikoa da sistema kudeaketa erraz baterako.

```bash
sudo apt update
sudo apt install -y wget gnupg
wget http://www.webmin.com/download/deb/webmin-current.deb
sudo apt install -y ./webmin-current.deb
```

Webmin zerbitzua instalatzean automatikoki abiarazten da. Egoera egiaztatzeko:

```bash
sudo systemctl status webmin
```

### Firewalla egokitzea

Ubuntu-n `ufw` erabiltzen baduzu, ireki `10000` portua:

```bash
sudo ufw allow 10000/tcp
sudo ufw reload
```

### Sarbidea

Nabigatzailean ireki:

- `https://<zure-ip-helbidea>:10000`


## 3. BIND instalatzea (DNS zerbitzaria)

- BIND instalatzeko, terminalean exekutatu:

    ```bash
    sudo apt install -y bind9 bind9utils bind9-doc dnsutils
    ```

- Oinarrizko konfigurazioa egin ebazten ez diren eskaerak ikastetxeko DNSra egiteko. Horretarako aldatu 

    ```bash
    nano named.conf.options
    ```

Hor aldatu:

````python
options {
    directory "/var/cache/bind";

    recursion yes;

    allow-recursion { localhost; 192.168.0.0/16; };
        allow-query-cache { localhost; 192.168.0.0/16; };

    forwarders {
        10.15.27.1;
    };
};
````

- Ondoren, zerbitzaria abiarazi eta hasierarekin batera martxan jarri:

    ```bash
    sudo systemctl enable --now named
    ```

- Suebakian dagokion portua ireki:

    ```bash
    sudo ufw allow 53/upd
    sudo ufw reload
    ```
