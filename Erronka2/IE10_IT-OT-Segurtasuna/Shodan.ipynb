# Shodan: Interneteko gailuen bilatzailea

Notebook honetan Shodanen funtzionamendua azalduko dugu: zer den, nola funtzionatzen duen, zein datu eskaintzen dituen eta nola erabil daitekeen segurtasun testetarako eta OT/IoT ingurunetarako.

## 1. Zer da Shodan?

- Shodan interneteko bilatzaile bat da, baina ez web orrialdeetarako; konektatutako gailu eta zerbitzarietarako izan ohi da.
- Webgune turistikoa bezala, Shodanek ordenagailu, kamerak, PLCak, SCADA sistemak eta IoT gailuak indexatzen ditu.
- GNU/Linux, Windows eta IoT gailuetatik zerbitzuen metadata eta publiko diren portuak bilduz funtzionatzen du.

Kontuan izan auekitzen ditugun gailu asko honeypotak izan daitezkela.

## 2. Zer nolako informazioa ematen du?

- IP helbidea eta portua
- Zerbitzu mota (http, ftp, ssh, telnet, modbus, mqtt, eta abar)
- Zerbitzariaren banner-a eta protokoloaren datuak
- HTTPS eta SSL/TLS ziurtagiriak
- Agian gailuaren eta aplikazioaren informazio nagusia

### Adibidez
- PLC bat aurkitu daiteke Modbus protokoloarekin eta azken konektatutako unea ikusi.
- Web-kamera publiko bat aurkitu daiteke portu 80 edo 8080ko HTTP zerbitzuarekin.

## 3. Nola egiten du bilaketa?

- Shodanek IP helbide askori eskaerak bidaltzen dizkie eta zerbitzuek bidaltzen dituzten erantzunak jasotzen ditu.
- Erantzunak analizatzen ditu eta zerbitzari bakoitzari etiketak, portuak eta protokoloak esleitzen dizkio.
- Ondoren, horiek indexatzen ditu kontsulta azkarretarako.

### Shodan bilaketa sintaxia
- `apache` - Apache zerbitzuak bilatzeko
- `port:22` - 22. portuan zabalik dauden gailuak
- `country:ES` - Espainian dauden gailuak
- `product:OpenPLC` - OpenPLC produktua duen zerbitzariak

## 4. Erabilera kasuak

### Segurtasun auditorietan
- Publiko dauden gailuen ikuspegi globala lortzeko
- Ahultasunak aztertzeko erabil daitezkeen zerbitzuen datuak
- OT/IoT sistemetan gutxieneko segurtasun-konfigurazio defizitak identifikatzeko

### Ikerketa eta monitorizazioa
- Zure sareko ikuspegia beste ikuspuntu batetik egiaztatzeko
- Fabrika edo eraikin bateko konektatutako gailuak identifikatzeko

### Arriskuak
- Publikoa eta segurtasunik gabeko gailuak aurkitu daitezke
- Informazio hori gaizki erabiltzen bada, sarrerako ateak ireki ditzake intrusioetarako

## 5. Shodan eta OT/IoT segurtasuna

- OT ingurunetan ere konektatutako gailu askok erabiltzen dituzte TCP/IP portuak.
- Shodanek Modbus, DNP3, IEC 60870-5-104 eta beste protokoloak bilatzen ditu.
- Segurtasunerako kontrolatzaileek eta OT operadoreek jakin behar dute zer gailu dauden publiko aurkitu daitezkeenak.

### Gomendatutako neurriak
- Sartu ot/IoT gailuak isolaturik edo zuzenean sarerik gabeko segmentu batean
- Segurtasun-iragazkiak eta sarbide kontrolak ezarri
- Publikoan eskuragarri ez dauden portuak itxi

## 6. Adibide praktikoa

1. Shodan webgunera joan eta kontu bat sortu (kontu gabe bilaketak oso mugatuta daude). Google kontuarekin lotuta sortzen bada datu gehiago gehitu beharrik ez dago.
2. Bilaketa bat egin, adibidez, `port:502`.
3. Aztertu emaitzak: IPak, portuak, eta zerbitzuen banner-ak.
4. Abian jarri segurtasun-neurriak ahultasun posibleak identifikatuta eta kalifikatu.

> Gogoan izan: Shodan bilaketak erabiltzeak beti jarraitu behar du lege eta etika egokiak, eta soilik zure sareko edo baimenarekin dituen sistemetan.
