# OpenPLC nola instalatu eta martxan jarri

OpenPLC PLC baten programazioa eta exekuzioa egiteko software sistema da. Bi zatitan banatzen dugu:
1. OpenPLC Runtime Docker container batean (guk horrela aukeratu dugulako).
2. OpenPLC Editor ordenagailuan instalatuta eta runtimearekin konektatzeko prest.

## Aurrebaldintzak

- OpenPLC runtime exekutatu behar den makinan Docker instalatuta egon behar da.

## 1. OpenPLC runtime abiarazi
Runtime martxan jartzeko, exekutatu komando hauek:

```python
!docker pull thiagoralves/openplc-runtime
```

```python
!docker run -d --name openplc-runtime -p 8080:8080 -p 502:502 thiagoralves/openplc-runtime
```

Zer egiten du komando honek?
- `-d`: atzeko planoan exekutatu.
- `--name openplc-runtime`: kontainerrari izen bat ematen dio.
- `-p 8080:8080`: web interfazearekin komunikatzeko atea zabaltzen du.
- `-p 502:502`: Modbus/TCP bezalako komunikazioetarako atea zabaltzen du.


### 1.1. Egiaztatu container-a martxan dagoela

```python
!docker ps
```

Erregistroak ikusteko:

```python
!docker logs -f openplc-runtime
```

### 1.2. Nola konektatu runtime-ra
Normalean, runtime web interfazeak `http://localhost:8080` helbidean eskuragarri egoten da. Hori nabigatzailean irekita, runtime martxan dagoela ikus dezakezu.



## 2. OpenPLC Editor instalatzea

OpenPLC Editor OpenPLC programak idazteko eta runtime-ra igotzeko erabiltzen da.

### 2.1. Deskargatu editorra
Aukeratu OpenPLC Editor-aren bertsio egokia eta deskargatu instalatzailea (https://autonomylogic.com/download):

![image](img/openplc-deskarga.png)

- Windows erabiltzeko, `.exe` instalatzailea ohikoena da.
- Linux erabiltzeko, AppImage erabiltzea gomendantzen da.

### 2.2. Instalatu Windows-en
1. Deskargatutako instalatzailea exekutatu.
2. Onartu baimenak eta jarraitu instalazio-urratsak.
3. Instalazioa amaitutakoan, OpenPLC Editor ireki.

### 2.3. Runtime-ra konektatu
Editorrean:
1. Aukeratu runtimearen IP helbidea.
2. Normalean, lokalean exekutatzen bada, erabil dezakezu `127.0.0.1` edo `localhost`.
3. Ezarri portua, normalean `8080` edo `502` erabiliz, araberakoa den konfigurazioaren arabera.
4. Konektatu eta probatu programak igotzeko.

## 3. Praktika honetan erabiliko den konfigurazioa
Gure kasuan, komenigarria da honakoa egitea:
- Docker container bat exekutatu runtime-rentzat.
- OpenPLC Editor instalatu eta konexioa ezarri `localhost:8080` edo `localhost` bidez.
- Programak idatzi eta runtime-ra igotzen saiatu.

## 5. Laburpena
OpenPLC runtime Docker-en exekutatzeko, lehenik Docker irudia deskargatu eta container bat abiarazi behar da. Ondoren, OpenPLC Editor instalatu eta runtimearekin konektatu. Horrela PLC programak eraiki eta probatu daitezke ingurune kontrolatuan.
