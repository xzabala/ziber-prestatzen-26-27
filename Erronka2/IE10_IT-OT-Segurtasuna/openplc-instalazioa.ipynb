## 1. OpenPLC runtime abiarazi
Runtime martxan jartzeko, exekutatu komando hauek:

```python
!docker run -d \
  --name openplc-v4 \
  --restart unless-stopped \
  -p 8443:8443 \
  -p 502:502 \
  ghcr.io/autonomy-logic/openplc-runtime:latest
```

Zer egiten du komando honek?
- `-d`: atzeko planoan exekutatu.
- `--name openplc-v4`: kontainerrari izen bat ematen dio.
- `--restart unless-stopped` : berrabiarazi gelditzen bada
- `-p 8443:8443`: web interfazearekin komunikatzeko atea zabaltzen du.
- `-p 502:502`: Modbus/TCP bezalako komunikazioetarako atea zabaltzen du.
