# Node-RED instalazioa Docker bidez

Notebook honetan Node-RED zer den eta nola instalatu Docker erabiliz azalduko da.

## Aurrebaldintzak

- Docker instalatuta egon behar da.
- Docker Engine martxan egon behar da.
- Portua 1880 libre egon behar da (Node-RED-ren lehenetsitako ataka).

```python
# Docker bertsioa egiaztatu
!docker --version
```

```python
# Node-RED irudia deskargatu
!docker pull nodered/node-red
```

```python
# Node-RED exekutatu Docker kontainerraren bidez
!docker run -d --name nodered -p 1880:1880 -v nodered_data:/data nodered/node-red
```

## Nola sartu

Instalazioa amaitutakoan, nabigatzailean honako helbidea ireki: http://localhost:1880

Hor ekingo duzu Node-RED editorera.

```python
# Kontainerra martxan dagoela egiaztatu
!docker ps --filter "name=nodered"
```

```python
# Kontaineraren log-ak ikusi
!docker logs nodered
```

## Optional: Docker Compose fitxategia

Hurrengo edukia `docker-compose.yml` fitxategi batean gorde daiteke eta `docker compose up -d` erabiliz abiarazi.

```python
# Docker Compose adibidea
compose_example = '''
services:
  nodered:
    image: nodered/node-red:latest
    container_name: nodered
    ports:
      - "1880:1880"
    volumes:
      - nodered_data:/data

volumes:
  nodered_data:
'''
print(compose_example)
```
