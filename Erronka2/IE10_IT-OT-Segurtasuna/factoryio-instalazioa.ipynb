# Factory I/O nola deskargatu eta instalatu

Factory I/O prozesu industrialen simulazioetarako softwarea da, simulazio industriala eraiki eta PLC/SCADA sistemekin konektatu ahal da.

## 1. Bertsio ezberdinak daude
Factory I/O-ren bertsio ezberdinak daude, eta horietako batzuk marka zehatzetarako edo protokolo zehatzetarako dira, adibidez:
- marka espezifikoetarako bertsioak (Siemens, Allen-Bradley, etab.)
- protokolo zehatzetarako bertsioak (Modbus, OPC UA, etab.)

Gure kasuan, ez dugu marka zehatz baten bertsioa behar; guk orokorra deskargatuko dugu eta instalatuko dugu, eta horrekin prozesu industriala simulatu eta komunikazioa ezartzeko aukera izango dugu.

<b> ADI: Simulatzaile honek GPUa erabiltzen du (ez du zertan oso ahaltsua izan), beraz ez da gomendagarria GPUrik gabeko makina birtualetan erabiltzea. GPUrik gabe 2 fps (frame segunduko) erakusten ditu eta simulazioek ez dute ondo erantzuten. </b>
